```python
# test_bkash_payment_page.py

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time

# --- CONFIGURATION ---
BASE_URL = "http://127.0.0.1:8000"
LOGIN_URL = "http://127.0.0.1:8000/login/"  # adjust if your login URL differs
BKASH_URL = "http://127.0.0.1:8000/payments/bkash/start/29/"  # adjust payment ID if needed
USERNAME = "investor_user1234"  # replace with valid investor username
PASSWORD = "TestPassword123"  # replace with valid investor password

# --- SETUP DRIVER ---
driver = webdriver.Chrome()
driver.maximize_window()
wait = WebDriverWait(driver, 10)

try:
    # --- Step 1: Login as Investor ---
    print("üîê Navigating to login page...")
    driver.get(LOGIN_URL)

    # Wait for login form to appear
    username_input = wait.until(EC.visibility_of_element_located((By.NAME, "username")))
    password_input = wait.until(EC.visibility_of_element_located((By.NAME, "password")))

    # Enter credentials
    username_input.send_keys(USERNAME)
    password_input.send_keys(PASSWORD)
    password_input.submit()
    print(f"‚úÖ Logged in successfully as '{USERNAME}'")
    time.sleep(2)

    # --- Step 2: Go to bKash payment simulation page ---
    driver.get(BKASH_URL)
    print("‚úÖ Opened bKash payment simulation page")
    time.sleep(2)

    # --- Step 3: Wait for the bKash buttons ---
    success_button = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".btn-success")))
    cancel_button = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".btn-danger")))
    print("‚úÖ Both 'Simulate success' and 'Cancel' buttons loaded successfully")
    time.sleep(2)

    # --- Step 4A: Test "Simulate Success" button ---
    success_button.click()
    print("üü¢ Clicked 'Simulate success' button")
    time.sleep(2)

    # --- Step 5A: Verify success message or redirect ---
    try:
        success_text = wait.until(
            EC.visibility_of_element_located(
                (By.XPATH, "//*[contains(text(), 'Success') or contains(text(), 'completed') or contains(text(), 'Payment successful')]")
            )
        )
        print("üéâ Success simulation verified: payment marked successful")
    except TimeoutException:
        print("‚ö†Ô∏è Success message not found ‚Äî verify redirect URL or success confirmation page")

    # --- Step 4B: Test "Cancel" button ---
    driver.get(BKASH_URL)
    cancel_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".btn-danger")))
    cancel_button.click()
    print("üî¥ Clicked 'Cancel' button")

    # --- Step 5B: Verify cancellation page or message ---
    try:
        cancel_text = wait.until(
            EC.visibility_of_element_located(
                (By.XPATH, "//*[contains(text(), 'Cancelled') or contains(text(), 'canceled') or contains(text(), 'Payment cancelled')]")
            )
        )
        print("‚úÖ Cancel simulation verified: payment cancelled successfully")
    except TimeoutException:
        print("‚ö†Ô∏è Cancel message not found ‚Äî verify cancel redirect or confirmation text")

    print("‚úÖ Full bKash payment simulation flow tested successfully after investor login!")

except TimeoutException as e:
    print(f"‚ùå Timeout waiting for element: {e}")
except Exception as e:
    print(f"‚ùå Unexpected error: {e}")
finally:
    
    print("üö™ Browser closed and test finished")
```
